```sage
from string import printable

def vigenere_decrypt(key, ciphertext, alphabet=printable):
    p = ''
    mod = len(alphabet)
    key_indices = [alphabet.index(k) for k in key]
    
    for i, char in enumerate(ciphertext):
        if char in alphabet:
            ind_c = alphabet.index(char)
            ind_k = key_indices[i % len(key)]
            p += alphabet[(ind_c - ind_k) % mod]
        else:
            p += char
    return p

# Példa futtatás
ciphertext = "RBABystSH"
key = "french"
plaintext = vigenere_decrypt(key, ciphertext)
print(plaintext)
```
