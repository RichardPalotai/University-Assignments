```python
#Ebben a cellában olvasd be a housing.csv fájlban található adatokat!

housing_data = pd.read_csv('housing.csv', index_col=0)
```

```python
print(housing_data.info)
housing_count = housing_data.shape[0]
```

#### 7. Határozd meg az oszloponkénti, majd a táblázatban szereplő összes hiányzó értéket! (1p)

```python
print(housing_data.isnull().sum())
print(housing_data.isnull().sum().sum())
```

#### 8. Tároljuk el, majd írassuk ki azoknak az oszlopoknak a nevét, ahol van hiányzó érték! (1p)

```python
columns_with_null = []
for col in housing_data.columns:
    if housing_data[col].isnull().sum() > 0:
        columns_with_null.append(col)
        print(col)

```

#### 9. Távolítsuk el helyben azokat az oszlopokat az adathalmazunkból, ahol az adatsornak legalább 15%-a hiányzik! (1p)

```python
# Használd segítségül a 4. és 8. feladat eredményeit!
for col in columns_with_null:
    if housing_data[col].isnull().sum()*100/housing_count >= 15:
        housing_data.drop(col, axis=1, inplace=True)
```

#### 10. Távolítsuk el helyben azokat az adatsorokat, ahol hiányzó érték található! Távolítsuk el helyben a PID oszlopot is! (1p)

```python
housing_data.dropna(inplace=True)
housing_data.drop("PID", axis=1, inplace=True)
```

#### 11. Ellenőrizzük le az adathalmazunkat!
#### Írassuk ki az oszlopok számát, adatsorok számát, hiányzó értékek számát és az alap információkat az adathalmazunkról! 
#### Mentsük el egy housing_cleaned nevű csv fájlba az eddigi formázásainkat! (0.5p)

```python
print(housing_data.shape[1])
print(housing_data.shape[0])
print(housing_data.isnull().sum().sum())
print(housing_data.info())
housing_data.to_csv('housing_cleaned.csv', index=True)
```

#### 12. Készítsünk egy "Has Fireplace" nevű oszlopot, amiben igaz-hamis értékeket tárolunk!
#### Az érték legyen igaz, ha a "Fireplaces" oszlop értéke több, mint nulla, egyébként legyen hamis! (1p) 

```python
housing_data["Has Fireplace"] = housing_data["Fireplaces"] > 0
```

#### 13. Ábrázoljuk a szám típusú adatok közötti összefüggést egy korrelációs mátrix segítségével!
#### Gondoskodjunk a megfelelő ábraméretről! (1.5p)

```python
correlation_matrix = housing_data.select_dtypes(include=['int64', 'float64']).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

#### 14. Rendezzük a korrelációs mátrix értékeit a "SalePrice" oszlop szerint növekvő sorrendbe!
#### Írassuk ki azoknak az oszlopoknak a neveit, amelyek szignifikánsan hozzájárulnak a "SalePrice" alakulásához (pl. 0.3-as küszöbbel)! (1.5p)

```python
correlation_matrix.sort_values(by="SalePrice", ascending=True, inplace=True)
print([col for col in correlation_matrix.columns if abs(correlation_matrix["SalePrice"][col]) > 0.3 and col != "SalePrice"])
```

#### 15. Készítsünk hisztogrammot a 2006-ban eladott házak árairól! ("Yr Sold", "SalePrice" oszlopok!) 
#### A hisztogrammot osszuk összesen 15 részre az adatvizualizációhoz. 
#### Címkézzük fel helyesen a tengelyeket, adjunk címet a grafikonnak és jelenítsük meg a grafikon rácsvonalait is! (1.5p)

```python
prices_2006 = housing_data[housing_data["Yr Sold"] == 2006]["SalePrice"]
plt.hist(prices_2006, bins=15, color='green', edgecolor='black')
plt.title("2006-os eladási árak eloszlása")
plt.xlabel("Eladási ár")
plt.ylabel("Gyakoriság")
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.show()
```

#### 16. Készítsünk egy ábrát két subplottal a 2006-ban és 2010-ben eladott telkekről!
##### A grafikonban az x tengelyen ábrázoljuk a telkek területét ("Lot Area" oszlop), az y tengelyen pedig ábrázoljuk a telkek árát ("SalePrice" oszlop).
##### Csak azokat a házakat vegyük figyelembe, ahol a telkek területe 6000 és 10000 négyzetláb között van!
##### A grafikonon a két különböző évben eladott telkek különböző színekkel, stackplottal, rendezetten kerüljenek jelölésre!
##### Lássuk el a grafikonjainkat címmel, címkékkel, illetve gondoskodjunk a megfelelő ábraméretről is! (2p)

```python
lots_2006 = housing_data[(housing_data["Yr Sold"] == 2006) & (housing_data["Lot Area"].between(6000, 10000, inclusive='neither'))][["Lot Area", "SalePrice"]]
lots_2006 = lots_2006.sort_values(by="Lot Area")
lots_2010 = housing_data[(housing_data["Yr Sold"] == 2010) & (housing_data["Lot Area"].between(6000, 10000, inclusive='neither'))][["Lot Area", "SalePrice"]]
lots_2010 = lots_2010.sort_values(by="Lot Area")
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(lots_2006["Lot Area"], lots_2006['SalePrice'], marker='o', color='blue')
plt.title("Eladási árak és telek terület 2006-ban")
plt.xlabel("Telek terület (négyzetláb)")
plt.ylabel("Eladási ár (Ft)")

plt.subplot(1, 2, 2)
plt.plot(lots_2010["Lot Area"], lots_2010['SalePrice'], marker='o', color='green')
plt.title("Eladási árak és telek terület 2010-ben")
plt.xlabel("Telek terület (négyzetláb)")
plt.ylabel("Eladási ár (Ft)")

plt.tight_layout()
plt.show()

```
